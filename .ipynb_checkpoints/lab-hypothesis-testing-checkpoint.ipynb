```python
#libraries
import pandas as pd
import scipy.stats as st
import numpy as np

```

```python
df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/pokemon.csv")
df
```

```python
df['Type 1'].value_counts()
```

```python
df['Type 2'].value_counts()
```

- We posit that Pokemons of type Dragon have, on average, more HP stats than Grass. Choose the propper test and, with 5% significance, comment your findings.

```python
#H0 = Dragons have more HP stats on average > Grass pokemons
```

```python
alpha = 0.05
```

```python
df_dragons = df[(df["Type 1"] == "Dragon") & (df["Type 2"] != "Grass")]["HP"]
df_grass = df[(df["Type 1"] == "Grass") & (df["Type 2"] != "Dragon")]["HP"]
```

```python
tstat, p_value = st.ttest_ind(df_dragons, df_grass, equal_var=False)
tstat, p_value
```

```python
if p_value >= alpha:
    print("We are not able to reject the null hypothesis")
else:
    print("We reject the null hypothesis")
```

We reject the null hypothesis.

Statistic (t-value):
The t-statistic is 3.33, which measures the difference between the sample means in units of standard error. A larger absolute value indicates a greater difference between the groups relative to the variability within the groups. In this case, a t-value of 3.33 suggests there is a noticeable difference between the two groups being compared.

p-value:
The p-value is 0.0016, which is the probability of observing a test statistic as extreme as the one calculated (or more extreme) if the null hypothesis is true. A smaller p-value indicates stronger evidence against the null hypothesis. Typically, if the p-value is less than a chosen significance level (commonly 0.05), we reject the null hypothesis. Here, since the p-value is much smaller than 0.05, it suggests that there is a statistically significant difference between the two groups being compared.

df (degrees of freedom):
The degrees of freedom is approximately 51.38. In t-tests, the degrees of freedom are used to determine the distribution the t-statistic follows, which influences the critical value needed for significance. The value 51.38 suggests that the sample size is reasonably large, allowing for more reliable conclusions.

Interpretation:
1) t-value = 3.33 suggests a notable difference between the two groups.
2) p-value of 0.0016 is quite low = indicating that the difference is statistically significant (under typical significance thresholds of 0.05 or 0.01).
3) degrees of freedom = 51.38, your sample size is sufficient to detect the difference accurately.

We consider a house (neighborhood) to be close to a school or hospital if the distance is lower than 0.50.

Hint:
- Write a function to calculate euclidean distance from each house (neighborhood) to the school and to the hospital.
- Divide your dataset into houses close and far from either a hospital or school.
- Choose the propper test and, with 5% significance, comment your findings.
 

```python
def euclidean_distance(lon1, lat1, lon2, lat2):
    return np.sqrt((lon1 - lon2) ** 2 + (lat1 - lat2) ** 2)
```

```python
school_lon = -118
school_lat = 34
hospital_lon = -122
hospital_lat = 37

# Define the lambda function to calculate Euclidean distance from the house to the school and hospital
df['Distance_to_School'] = df.apply(lambda row: (euclidean_distance(row['longitude'], row['latitude'], school_lon, school_lat)), axis=1)
df['Distance_to_Hospital'] = df.apply(lambda row: (euclidean_distance(row['longitude'], row['latitude'], hospital_lon, hospital_lat)), axis=1)
# Print the DataFrame with the calculated distances
df
```

```python
df['close'] = (df['Distance_to_School']< 0.50) | (df['Distance_to_Hospital'] < 0.50)
df
```
